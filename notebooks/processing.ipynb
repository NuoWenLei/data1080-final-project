```python
import numpy as np, pandas as pd
```

```python
with open("../tickers/tickerNumerics.json", "r") as num_json:
    numerics = json.load(num_json)
```

```python
with open("../tickers/tickerMetadata.json", "r") as meta_json:
    metadata = json.load(meta_json)
```

```python
numeric_df = pd.DataFrame(numerics, index = [t["symbol"] for t in metadata])
numeric_df.shape
```

```python
numeric_df.head()
```

```python
numeric_df[numeric_df == np.inf] = np.nan
numeric_df[numeric_df == -np.inf] = np.nan
```

```python
attribute_nan_pct = numeric_df.isna().sum(axis = 0) / numeric_df.shape[0]
```

```python
# filter out attributes that are missing for more than half of the tickers
attrs_to_drop = attribute_nan_pct[attribute_nan_pct > .5].index.values
attrs_to_drop
```

```python
numeric_df = numeric_df.drop(attrs_to_drop, axis = 1)
numeric_df.shape
```

```python
rank_df = pd.DataFrame([], index = numeric_df.index)
```

```python
for col in numeric_df.columns:
    rank_df[col] = numeric_df[col].rank(pct = True)
```
